# SIRD Model

Importe benötigt um das SIRD Modell zu bentzen.

```python
import networkx as nx
from IPython.display import HTML
from SIRD.simulation import Simulation
```

Initialisierung aller benötigten Größen

```python
n_agents = 144
n_init_infected = 2
n_steps = 60
beta = 0.35
mu = 0.1
mu_d = 0.08 
```

Erstellung des Kontaktnetzwerks mit Networkx

```python
contact_network = nx.watts_strogatz_graph(n_agents, k=4, p=0.03)
#contact_network = nx.erdos_renyi_graph(n_agents, p=1)
#contact_network = nx.barabasi_albert_graph(n_agents, m=1)
```

Initialiserung der Simulation

```python
sim = Simulation(
    n_steps, 
    beta, 
    mu, 
    mu_d, 
    n_agents, 
    n_init_infected,  
    video=True,
    contact_network=contact_network
)
```

Mit der run() Methode wird simuliert.

```python
sim.run()
```

Es gibt eine Methode, um die Ergebnisse zu plotten.

```python
sim.show_results()
```

Es gibt eine Methode um eine animation zu erstellen

```python
video = sim.generate_video(show=False)
HTML(video.to_html5_video())
```
